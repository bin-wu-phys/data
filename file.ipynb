```python
latt = Lattice(('r tilde', 0.0, 3.0, 31),
                ('vz tilde', 0.0, 10.0, 21),
                ('phir', 10),
                ('theta', 10))
```

# 1. Initialize kt

```python
t0 = 0.1
dt = 0.2
dr = 0.1
ob = Observable(t0, latt)
ic = InitCond('gaussian', t0, (2, 1.0))
C = Kernel(ob=ob, dr=dr)
kt = KinTran(t0, ic.gauss(latt), C, dt=dt)
```

```python
kt.calcvn()
```

```python
kt.kern.ob.interp(kt.F)
```

```python
nr = 5
nph = 5
print(kt.kern.ob.Finterp[2][0][0](kt.kern.ob.latt.lattice[kt.kern.ob.idx_r][nr], kt.kern.ob.latt.lattice[kt.kern.ob.idx_ph][nph])[0]
      +1.0j*kt.kern.ob.Finterp[2][0][1](kt.kern.ob.latt.lattice[kt.kern.ob.idx_r][nr], kt.kern.ob.latt.lattice[kt.kern.ob.idx_ph][nph])[0], 
      kt.F[2, 0, nr, nph])
```
